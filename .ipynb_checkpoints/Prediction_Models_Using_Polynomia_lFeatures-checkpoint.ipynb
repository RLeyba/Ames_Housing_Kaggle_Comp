```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import *
%matplotlib inline
```

```python
#read in training data and testing data
data = pd.read_csv('./datasets/cleaned.csv', index_col = 'Unnamed: 0')
test_final = pd.read_csv('./datasets/test_final.csv', index_col = 'Unnamed: 0')
```

```python
#set target data
y = data['SalePrice']
```

```python
#check to see if data is skewed. Sales price is right skewed. 
#Could have potential to take the log of the data to fix the outliers
data['SalePrice'].hist()
```

## Feature Engineering my data 

```python
#set training data
X = data[['Lot Area',
          'Overall Qual',
          'Year Built',
          'Year Remod/Add',
          'Mas Vnr Area',
          'Exter Qual',
          'Bsmt Qual', 
          'Gr Liv Area',
          'Full Bath',
          'Half Bath',
          'Total Bsmt SF',
          'Garage Cars',
          'Garage Area',
          'Fireplaces',
          'TotRms AbvGrd',
          'Kitchen Qual'
         ]] 
```

```python
#set test data
test = test_final[['Lot Area',
          'Overall Qual',
          'Year Built',
          'Year Remod/Add',
          'Mas Vnr Area',
          'Exter Qual',
          'Bsmt Qual', 
          'Gr Liv Area',
          'Full Bath',
          'Half Bath',
          'Total Bsmt SF',
          'Garage Cars',
          'Garage Area',
          'Fireplaces',
          'TotRms AbvGrd',
          'Kitchen Qual'
         ]] 
```

```python
#check if training data, target date, and test data have the correct shape.
#Target and training have quivalent rows. Test and trainging have consistent columns.
print('target shape: ', y.shape)
print('features shape: ', X.shape)
print('test shape: ', test.shape)
```

```python
#Feature engineering: get polynomial features from the training and testing data I have.
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
# Fit and transform our X data.
Xoverfit = poly.fit_transform(X)
test_overfit = poly.fit_transform(test)
```

```python
#set new features to a new data frame
poly_df = pd.DataFrame(columns = poly.get_feature_names(X.columns), data =Xoverfit  )
```

```python
poly_df.head()
```

```python
# get columns that only have a correlation to sale price over .6. 
#I played around and found .6 gave me the r2 scores with my model and data

#appened sale price to poly_df to find correlations with selected columns
poly_df['SalePrice'] = data['SalePrice']

#sort values by correlation and mask values that are below .6
best_corr = poly_df.corr()['SalePrice'].sort_values(ascending=False) > 0.6

#appenend columns names that are above .6 to later use as an index on our training and testing data
good_columns = []
for i in best_corr.index:
    if best_corr[i] == True:
        good_columns.append(i)
```

```python
#see good columns
good_columns
```

```python
#trim the fat and only use the good columns and drop sale price since we do not need it anymore
X_trim = poly_df[good_columns]
X_trim = X_trim.drop('SalePrice', axis=1)
```

```python
#check shape
X_trim.shape
```

```python
#remove sale price form good columns so You can index your testing data now
good_columns.remove('SalePrice')
```

```python
#first create a data frame with your new array of polynomial feature testing data
test_trim = pd.DataFrame(columns = poly.get_feature_names(test.columns), data =test_overfit )
```

```python
#index the data frame with the good columns
test_trim = test_trim[good_columns]
```

```python
#check shape to see if it matches your training data columns
test_trim.shape
```

## Instatiating and fitting my Predictive models

```python
# Instantiate Standard Scaler.
ss = StandardScaler()
# Fit AND transform Standard Scaler to overfit data in one step.
X_sc = ss.fit_transform(Xoverfit)
test_sc = ss.transform(test_overfit)
```

```python
X_overfit_train, X_overfit_test, y_overfit_train, y_overfit_test = train_test_split(X_sc,
                                                                                    y,
                                                                                    test_size=0.35,
                                                                                    random_state=42)
```

```python
print(f'X_overfit_train shape is: {X_overfit_train.shape}')
print(f'y_overfit_train shape is: {y_overfit_train.shape}')
print(f'X_overfit_test shape is: {X_overfit_test.shape}')
print(f'y_overfit_test shape is: {y_overfit_test.shape}')
```

```python
X_overfit_train.mean()
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_overfit_train, y_overfit_train)
```

```python
lr_preds = lr_model.predict(X_overfit_test)
```

```python
lr_preds
```

```python
lr_model.coef_
```

```python
r2_score(y_overfit_test, lr_preds)
#83.0% of the variability in our 
#(wine quality) is explained by our model
#relative to a baseline model with no predictors.
```

```python
lr_preds_train = lr_model.predict(X_overfit_train)
r2_score(y_overfit_train, lr_preds_train)

#overfit since its better training is better than testing, need to regularize
```

```python
test_lr_preds= lr_model.predict(test_sc)
```

```python
test_lr_preds
```

```python
# Instantiate.
ridge_model = Ridge(alpha=10)

# Fit.
ridge_model.fit(X_overfit_train, y_overfit_train)

# Generate predictions.
ridge_preds = ridge_model.predict(X_overfit_test)
ridge_preds_train = ridge_model.predict(X_overfit_train)

# Evaluate model using R2.
print(r2_score(y_overfit_train, ridge_preds_train))
print(r2_score(y_overfit_test, ridge_preds))
```

```python
from sklearn.linear_model import RidgeCV
```

```python
# Set up a list of ridge alphas to check.
r_alphas = np.logspace(0, 5, 100)
# Generates 100 values equally between 0 and 5,
# then converts them to alphas between 10^0 and 10^5.

# Cross-validate over our list of ridge alphas.
ridge_model = RidgeCV(alphas=r_alphas, scoring='r2', cv=5)

# Fit model using best ridge alpha!
ridge_model.fit(X_overfit_train, y_overfit_train)
```

```python
ridge_model.alpha_
```

```python
y_pred_train_ridge = ridge_model.predict(X_overfit_train)
y_pred_test_ridge = ridge_model.predict(X_overfit_test)

print(r2_score(y_overfit_train, y_pred_train_ridge))
print(r2_score(y_overfit_test, y_pred_test_ridge))
```

```python
test_ridge_pred = ridge_model.predict(test_sc)
```

```python
test_ridge_pred
```

```python
from sklearn.linear_model import LassoCV
```

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace(-3, 0, 100)

# Cross-validate over our list of Lasso alphas.
lasso_model = LassoCV(alphas=l_alphas, cv=5)

# Fit model using best ridge alpha!
lasso_model = lasso_model.fit(X_overfit_train, y_overfit_train)
```

```python
# Here is the optimal value of alpha
lasso_model.alpha_
```

```python
# Generate predictions
y_pred_train_lasso = lasso_model.predict(X_overfit_train)
y_pred_test_lasso = lasso_model.predict(X_overfit_test)

# Evaluate model.
print(r2_score(y_overfit_train, y_pred_train_lasso))
print(r2_score(y_overfit_test, y_pred_test_lasso))
```
